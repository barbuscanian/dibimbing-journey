# **Lakukan data manipulation pada jumlah_bus dan jumlah_penumpang supaya berada dalam satu format type**

```python
combined_Penumpang['jumlah_penumpang'] = combined_Penumpang['jumlah_penumpang'].astype(str)
combined_Penumpang['jumlah_penumpang'] = combined_Penumpang['jumlah_penumpang'].str.replace(',', '', regex=True).replace(['nan', 'inf', '-inf'], '0', regex=True)
combined_Penumpang['jumlah_penumpang'] = combined_Penumpang['jumlah_penumpang'].astype(float).round().astype(int)
```

```python
combined_Penumpang['jumlah_bus'] = combined_Penumpang['jumlah_bus'].astype(str)
combined_Penumpang['jumlah_bus'] = combined_Penumpang['jumlah_bus'].str.replace(',', '', regex=True).replace(['nan', 'inf', '-inf'], '0', regex=True)
combined_Penumpang['jumlah_bus'] = combined_Penumpang['jumlah_bus'].astype(float).round().astype(int)
```

```python
print(combined_Penumpang)
```

# **Lakukan aggregasi sum, mean, dan max dari jumlah_bus dan jumlah_penumpang berdasarkan tahun**

```python
aggregated_data = combined_Penumpang.groupby('tahun').agg({'jumlah_bus': ['sum', 'mean', 'max'], 'jumlah_penumpang': ['sum', 'mean', 'max']})
aggregated_data.columns = ['total_bus_sum', 'total_bus_mean', 'total_bus_max',
                           'total_penumpang_sum', 'total_penumpang_mean', 'total_penumpang_max']

print(aggregated_data)
```


# **Simpan hasil aggregasi dalam ZIP  berisi (json file dan .ipynb)**

```python
result_list = []
for year, group in aggregated_data.iterrows():
    year_data = {
        'tahun': int(year),
        'jumlah_bus': {
            'sum': int(group['total_bus_sum']),
            'mean': float(group['total_bus_mean']),
            'max': int(group['total_bus_max'])
        },
        'jumlah_penumpang': {
            'sum': int(group['total_penumpang_sum']),
            'mean': float(group['total_penumpang_mean']),
            'max': int(group['total_penumpang_max'])
        }
    }
    result_list.append(year_data)

with open('hasil_agregasi.json', 'w') as json_file:
    json.dump(result_list, json_file, indent=4)
```
